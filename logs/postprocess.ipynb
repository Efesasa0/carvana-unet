```python
import glob
import json
import matplotlib.pyplot as plt

json_files = glob.glob("./*.json")
plt.figure(figsize=(10, 6))

for file in json_files:
    with open(file, 'r') as fp:
        data = json.load(fp)
        train_losses = data[0]
        validation_losses = data[1]
        #plt.plot(train_losses, label=f"{file} - train")
        plt.plot(validation_losses, label=f"{file} - val", alpha=0.6,linewidth=2.0
                 )

plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.title("Validation Losses | img_shape-feat-batch_size-epochs-lr")
plt.legend()
plt.ylim(0, 0.25)
plt.show()
```
